```python
import pandas as pd
import numpy as np
```

```python
filepath = "/home/jouell/code/jouell3/NBA_app/Sport_stat/input/Player Totals.csv"
df_players = pd.read_csv("data/Player Totals.csv")
df_team_abbrev = pd.read_csv('data/Team Abbrev.csv')
len(df_players)
```

```python
def abbreviation(team):
    if team in df_team_abbrev['abbreviation'].values:
        return df_team_abbrev[df_team_abbrev['abbreviation'] == team]['team'].values[0]
    else:
        return None
```

```python
df_players["full_team_name"] = df_players["team"].apply(abbreviation)
df_players
```

```python
with open("data/Player Totals.csv", "w"):
     df_players.to_csv("data/Player Totals.csv", index=False)
```
