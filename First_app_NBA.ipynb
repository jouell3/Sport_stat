```python

def strip_accents(s):
    if isinstance(s, str):
        return ''.join(
            c for c in unicodedata.normalize('NFD', s)
            if unicodedata.category(c) != 'Mn'
        )
    return s


def add_team(player):
    player_striped = strip_accents(player)
    selected_player = df_players[(df_players["player_stripped"] == player_striped)]
    if len(selected_player) > 0:
        return selected_player.iloc[0]["full_team_name"]
    else:
        return None
```

```python
def fill_team_full_name(player_ID, year):
    """
    Find team name for a player in a specific year using fuzzy matching
    
    Args:
        player_ID: Player ID from player_salaries
        year: Year to search
        
    Returns:
        team_full_name from df_players or None if not found
    """
    # Filter df_players for the specific year
    year_data = df_players[df_players["season"] == int(year)].copy()
    if len(year_data) == 0:
        return None
    
    # match player_ID with ID in df_players
    exact_match = year_data[year_data["player_id"] == player_ID]
    if len(exact_match) > 0:
        return exact_match.iloc[0]["full_team_name"]
    
    return None
```

```python
def final_try(player, year):
    """
    Find team name for a player in a specific year
    
    Args:
        player: Player name from player_salaries
        year: Year to search
        
    Returns:
        team_full_name from df_players or None if not found
    """
    year_data = df_players[df_players["season"] == int(year)].copy()
    
    if len(year_data) == 0:
        return None
    
    names = strip_accents(player).split()
    if len(names) > 1:
        # Search by last name (second element)
        last_name = names[1]
        matches = year_data[year_data["player_stripped"].str.contains(last_name, case=False, na=False)]
        
        if len(matches) == 0:
            return None
        elif len(matches) == 1:
            return matches.iloc[0]["full_team_name"]
        else:
            # Multiple matches, filter by first name
            first_name = names[0]
            refined_matches = matches[matches["player_stripped"].str.contains(first_name, case=False, na=False)]
            
            if len(refined_matches) == 1:
                return refined_matches.iloc[0]["full_team_name"]
            elif len(refined_matches) > 1:
                # Still multiple matches, return the first one
                return refined_matches.iloc[0]["full_team_name"]
            else:
                return None
    else:
        # Only one name provided
        matches = year_data[year_data["player"].str.contains(names[0], case=False, na=False)]
        if len(matches) > 0:
            return matches.iloc[0]["full_team_name"]
        
    return None
```

```python
df_salaries['full_team_name'] = df_salaries.apply(lambda row: final_try(row["player"], row["year"]), axis=1)
df_salaries.isna().sum()
```
