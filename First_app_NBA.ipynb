```python
team_salaries = pd.DataFrame(columns=["team", "sum_salary", "today", "year"])
for file in files:
    df = pd.read_csv(file, header=0, names=["Ranking", "team", "sum_salary", "today"])
    df["year"] = file.name[:4]
    df = df.drop("Ranking", axis=1).reset_index(drop=True)
    team_salaries = pd.concat([team_salaries, df], ignore_index=True)
team_salaries
```

```python
team_salaries.to_csv("data/team_salaries.csv", index=False)
```
